```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_df=pd.read_csv(city_data_to_load)
ride_df=pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
city_df = city_df.drop_duplicates('city', keep = 'first')

merged_df=city_df.merge(ride_df, on = 'city', how = 'outer')


# Display the data table for preview
merged_df.head()

```

## Bubble Plot of Ride Sharing Data

```python
urban_df = merged_df.loc[(merged_df["type"] == "Urban")]
suburban_df = merged_df.loc[(merged_df["type"] == "Suburban")]
rural_df = merged_df.loc[(merged_df["type"] == "Rural")]

# Obtain the x and y coordinates for each of the three city types
#Urban cities
avg_fare_urban = urban_df.groupby(['city'])['fare'].mean()
total_rides_urban = urban_df.groupby(['city']).count()['ride_id']
total_drivers_urban= urban_df.groupby(['city'])['driver_count'].value_counts()

#Suburban cities
avg_fare_suburban = suburban_df.groupby(['city'])['fare'].mean()
total_rides_suburban = suburban_df.groupby(['city']).count()['ride_id']
total_drivers_suburban = suburban_df.groupby(['city'])['driver_count'].value_counts()

#Rural cities
avg_fare_rural = rural_df.groupby(['city'])['fare'].mean()
total_rides_rural = rural_df.groupby(['city']).count()['ride_id']
total_drivers_rural = rural_df.groupby(['city'])['driver_count'].value_counts()

# Build the scatter plots for each city types
plt.scatter(total_rides_urban, avg_fare_urban,s=total_drivers_urban*10,
            marker ='o', facecolors ="lightcoral", edgecolors='black',alpha = 0.5, label="Urban")

plt.scatter(total_rides_suburban, avg_fare_suburban,s=total_drivers_suburban*10,
            marker ='o', facecolors ="lightskyblue", edgecolors='black',alpha = 0.5, label="Suburban")

plt.scatter(total_rides_rural, avg_fare_rural,s=total_drivers_rural*10,
            marker ='o', facecolors ="gold", edgecolors='black',alpha = 0.5, label="Rural")
# Incorporate the other graph properties
plt.title("Pyber Ride Sharing Data (2016)")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
plt.grid()
# Create a legend
lgnd= plt.legend(scatterpoints=1, frameon=False, labelspacing=1, title='City Type')



# Incorporate a text label regarding circle size
plt.text(44, 30,"Note: Circle size correlates with driver count per city.")

print(f"Based on the data I observed Urban fares average lower than Suburban but more rides are needed. To further analyze the Urban locations would probably benefit from an increase in employees in these areas." )
# Save Figure
plt.savefig("Images/Pyber Ride Sharing.png")
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
total_city_fare=merged_df.groupby(['type'])['fare'].sum()
# Build Pie Chart
labels = total_city_fare.index
sizes = total_city_fare
colors = ['gold', 'lightskyblue', 'lightcoral']
explode = (0, 0, 0.1) 
# Plot
plt.title("% of Total Fares By City Types")

plt.pie(total_city_fare, explode= explode, labels=labels, colors=colors, autopct="%1.1f%%", shadow=True, startangle=160)
plt.axis('equal')

print(f"By an overwhelming 62.7% Urban fares have the largest number of fares which is closely related to the total number of rides for Urban. This might have to do with the fact that in most urban setings a personal car isn't necessary. The short distanes make it lesdesirable to pay for the aspects of owning a vechicle vs ride share prices." )
 
plt.show()
# Save Figure
plt.savefig("Images/Total Fares By City Types.png")
```

## Total Rides by City Type

```python
# Calculate Ride Percents

# Build Pie Chart

# Save Figure


# Calculate Type Percents
total_rides=merged_df.groupby(['type'])['ride_id'].sum()
# Build Pie Chart
labels = total_rides.index
sizes = total_rides
colors = ['gold', 'lightskyblue', 'lightcoral']
explode = (0, 0, 0.1)  # explode 1st slice
 
# Plot
plt.title("% of Total Rides By City Types")

plt.pie(total_rides, explode= explode, labels=labels, colors=colors, autopct="%1.1f%%",shadow=True, startangle=160)
plt.axis('equal')
print(f" With only 5% of rides coming from the rural type, its possible to also believe that most drivers can not fair being out that far for just a few trips. Its possible to make more trips and more money vs a couple trips for further distances.")
plt.savefig("Images/Total Rides By City Types.png")
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Driver Percents

# Build Pie Charts

# Save Figure


# Calculate Type Percents
total_drivers=merged_df.groupby(['type'])['driver_count'].sum()
# Build Pie Chart
labels = total_drivers.index
sizes = total_drivers
colors = ['gold', 'lightskyblue', 'lightcoral']
explode = (0, 0, 0.1)  # explode 1st slice
 
# Plot
plt.title("% of Total Drivers By City Types")

plt.pie(total_drivers, explode= explode, labels=labels, colors=colors, autopct="%1.1f%%",shadow=True, startangle=160)
plt.axis('equal')
print(f"Looking at the other graphs, the small percentage of drivers from the rural area reflect the traveling distance from rural to urban. While it may be more pleasing to make more money, a rural driver might have to also commute that far for day-to-day work which could casue more wear and tear for a personal vechicle thus making the idea of ride sharing more expensive. If a person already resides in  the urban area it would also be easier to navigate a well known area too.")
plt.savefig("Images/Total Drivers By City Types.png")
plt.show()
```
